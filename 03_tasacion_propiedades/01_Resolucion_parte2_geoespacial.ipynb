## 1) Trabajo con CABA

Generamos una máscara que incluya a los datos de propiedades de CABA y Buenos Aires para poder graficarlos en mapas más detallados, ya que son la amplia mayoría de los registros del dataset original

```python
conditions = [
    'Capital Federal',
    'Bs.As. G.B.A. Zona Norte',
    'Bs.As. G.B.A. Zona Sur',
    'Bs.As. G.B.A. Zona Oeste',
    'Buenos Aires Interior',
    'Buenos Aires Costa Atlántica'
]

mask_buenos_aires = geo_data['state_name'].isin(conditions)
geo_data_bsas = geo_data[mask_buenos_aires]

geometria_bsas = gpd.points_from_xy(geo_data_bsas.lon_clean, geo_data_bsas.lat_clean)
geo_df_data_bsas = gpd.GeoDataFrame(geo_data_bsas, geometry=geometria_bsas)

geo_df_data_bsas.head(3)
```

Finalmente, graficamos las propiedades que se encuentran dentro del polígono de la Ciudad Autónoma de Buenos Aires

```python
barrios = "C:/Users/Tefi/OneDrive/Documentos/GitHub/Proyectos_DH/03_tasacion_propiedades/barrios.csv"

df_barrios = pd.read_csv(barrios, encoding='latin1')

df_barrios["WKT"] = df_barrios["WKT"].apply(shapely.wkt.loads) 
geo_barrios = gpd.GeoDataFrame(df_barrios, geometry='WKT')
geo_barrios.sample(10)
```

```python
# Búsqueda de puntos dentro de los límites de los barrios
from geopandas.tools import sjoin
merged = sjoin(geo_df_data_bsas, geo_barrios, how='inner', op='within')

# Columna 'geometry_clean' con los valores que se encuentran dentro de los límites de los barrios
geo_df_data_bsas['geometry_clean'] = merged['geometry']

geo_df_data_bsas_clean = geo_df_data_bsas['geometry_clean']

geo_df_data_bsas_clean.head()
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
geo_barrios.plot(ax=ax, color='white', edgecolor='black')

geo_df_data_bsas_clean.plot(ax=ax, color='red', markersize=5)

plt.title('Propiedades en CABA')
plt.show()
```

Revisamos cual es la distribución de las viviendas en el Área Metropolitana de Buenos Aires por tipo de vivienda.

```python
# Lista de state_names de interés
estados_interes = ['Capital Federal','Bs.As. G.B.A. Zona Norte', 'Bs.As. G.B.A. Zona Sur', 'Bs.As. G.B.A. Zona Oeste']

# Número de gráficos por fila
graficos_por_fila = 2

mi_paleta =["#55AA77","#80D8FF", "#BF80FF" ,"#FF99B3"] 

# Calcular el número total de filas necesarias
num_filas = -(-len(estados_interes) // graficos_por_fila)  # Redondeo hacia arriba usando la división entera

# Crear subgráficos
fig, axes = plt.subplots(num_filas, graficos_por_fila, figsize=(15, 5 * num_filas))
axes = axes.flatten()  # Aplanar la matriz de subgráficos

# Crear un gráfico de torta para cada estado
for i, estado in enumerate(estados_interes):
    # Filtrar el DataFrame solo para el estado actual
    gdf_estado = geo_df_data[geo_df_data['state_name'] == estado]

    # Agrupar por 'state_name' y 'property_type' y contar las propiedades en cada grupo
    property_count = gdf_estado.groupby(['state_name', 'property_type']).size()

    # Convertir el resultado en un DataFrame y restablecer el índice
    property_count_df = property_count.reset_index(name='property_count')

    # Ordenar el DataFrame por la columna 'property_count' de mayor a menor
    property_count_df = property_count_df.sort_values(by='property_count', ascending=False)

    # Preparar los datos para el gráfico de torta
    labels = property_count_df['property_type']
    sizes = property_count_df['property_count']
    colors = [plt.cm.Set3(j / len(property_count_df)) for j in range(len(property_count_df))]  # Colores escalados según el índice

    # Graficar el gráfico de torta en el subgráfico correspondiente
    ax = axes[i]
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=mi_paleta )
    ax.axis('equal')  # Así se ve más como un círculo
    ax.set_title(f'Distribución en {estado}')

# Ajustar el diseño y mostrar los gráficos
plt.tight_layout()
plt.show()
#, 
```

A continuación, ubicamos los departamentos en el mapa de la Provincia de Buenos Aires para que sea mas visible gráficamente.
